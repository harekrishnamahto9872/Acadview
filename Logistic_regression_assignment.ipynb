```
df.drop(['Survived'],axis=1,inplace=True)
```

```
df['Cabin'].head()
```

```
def fun(x):
  if pd.isnull(x):
    return(0)
  else:
    return(1)
```

##Creating a new column 'Has_cabin' with binary value 0 or 1.

```
df['Has_Cabin']=df['Cabin'].apply(fun)
```

```
df.columns
```

```
df.drop(['Cabin'])
```

```
df.columns
```

```
df.head()
```

##Extracting title using lambda expression and apply() method

```
import re
df['Title'] = df.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
```

```
df['Title'].head()
```

##Replacing Titles

```
df['Title']=df['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
```

```
df['Title']=df['Title'].replace(['Don', 'Dona', 'Rev', 'Dr','Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')
```

```
df.isnull().sum()
```

```
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```
df.isnull().sum()
```

##Creating two new columns using qcut() pandas method 

```
df['Categorical_Age']=pd.qcut(df['Age'],5,labels=[0,1,2,3,4])
df['Categorical_Fare']=pd.qcut(df['Fare'],5,labels=[0,1,2,3,4])
```

```
df.head()
```

```
df.columns
```

```
df.head()
```

##Converting categorical data of 'Sex' and 'Embarked' into numerical using pandas.get_dummies

```
gender=pd.get_dummies(df['Sex'])
```

```
embark=pd.get_dummies(df['Embarked'])
```

```
gender.head()
```

##Concatinating all dataframes

```
df=pd.concat([df,gender,embark],axis=1)
```

##Dropping unrequired columns

```
df.drop(['Name','PassengerId','Ticket','Age','Fare','SibSp','Parch','Title','Embarked','Sex','Cabin'],axis=1,inplace=True)
```
