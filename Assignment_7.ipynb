```
sns.boxplot(x='Age',data=df)
```

1.   The Median of ages of people is near to 30.
2.   Majority of the people have ages between 15 to 40.



```
sns.countplot(df['Survived'])
```

1.We can see that majority of the people died while only few people survived on borad.   

```
sns.barplot(x='Sex',y='Survived',data=df)
```

The above plot shows that female survivals  were more than male survival.

```
sns.distplot(df['Age'],kde=False)
```

The above plot shiws that most of the people were having age in between 15-40.

```
sns.factorplot(x='Survived',hue='Sex',kind='count',data=df)
```

```
sns.factorplot(x='Pclass',data=df,kind='count')
```

The above plot shows that Pclass 3 is the highest among all the classes.

```
sns.factorplot('Survived',hue='Pclass',data=df,kind='count')
```

The above plot shows survival of the members grouped by Pclass.

```
sns.factorplot('Pclass',hue='Sex',data=df,kind='count')
```

The above plot shows the count of persons in individual Pclass grouped by Gender.

```
df.corr()
```

The above table shows the co-relation between all pairs of columns.

```
sns.heatmap(df.corr())
```

1.   There is a strong Co-relation between "Fare " and "Pclass".

```
sns.regplot(x='Fare',y='Pclass',data=df,x_jitter=0.2)
```

1.   The above plot shows that the higher the class of person ,the higher is the fare..
2.   High Class person have high Fare.

1.   There is linear relationship between Fare and Person Class.









```
sns.regplot(x='Age',y='Siblings/Spouses Aboard',data=df)
```



1.   The above plot shows person with more age are having Siblings/Spouses Abroad.
